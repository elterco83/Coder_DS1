Modelo de Predicción de Ventas
- Objetivo: Predecir las ventas de productos Adidas en función de variables como la temporada, la región, el precio
- Variables predictoras: Temporada, región, precio, método de venta, producto, etc.
- Variable objetivo: Ventas totales
- Técnica: Regresión lineal múltiple o modelos de series de tiempo



```python
import pandas as pd

# Convertir la columna 'fecha' a datetime
df_base['fecha'] = pd.to_datetime(df_base['fecha'], errors='coerce')

# Función para determinar la temporada a partir del mes
def obtener_temporada(fecha):
    mes = fecha.month
    if mes in [12, 1, 2]:
        return 'Invierno'
    elif mes in [3, 4, 5]:
        return 'Primavera'
    elif mes in [6, 7, 8]:
        return 'Verano'
    elif mes in [9, 10, 11]:
        return 'Otoño'

# Crear una nueva columna 'temporada' extrayendo la estación de la fecha
df_base['temporada'] = df_base['fecha'].apply(lambda x: obtener_temporada(x) if pd.notnull(x) else None)

# Convertir la variable 'temporada' a dummies (opcional: drop_first=True para evitar la multicolinealidad)
df_base = pd.get_dummies(df_base, columns=['temporada'], drop_first=True)

# Si ya no necesitas la columna 'fecha', puedes eliminarla
df_clean = df_base.drop(columns=['fecha'])
```

```python
#Convertir variables categóricas a dummies
categoricas = ['ciudad','estado','sexo', 'retailer', 'region', 'metodo de venta', 'producto']
df_clean = pd.get_dummies(df_clean, columns=categoricas, dtype=int)
```

```python
import pandas as pd

# Supongamos que tu DataFrame original se llama 'df'
# y que la variable objetivo es 'ventas totales'

# Calcular los cuartiles y el IQR
Q1 = df_clean['ventas totales'].quantile(0.25)
Q3 = df_clean['ventas totales'].quantile(0.75)
IQR = Q3 - Q1

# Definir los límites inferior y superior
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filtrar el DataFrame para eliminar outliers en 'ventas totales'
df_final = df_clean[(df_clean['ventas totales'] >= lower_bound) & (df_clean['ventas totales'] <= upper_bound)]

print("Tamaño original:", df_base.shape)
print("Tamaño después de eliminar outliers:", df_final.shape)

# Ahora puedes continuar usando df_clean para el resto de tu análisis (por ejemplo, stepwise, entrenamiento, etc.)
```

```python
X = df_final.drop(columns=['ventas totales'])  # mi variable objetivo
y = df_final['ventas totales']
```

```python
import pandas as pd
import statsmodels.api as sm

def stepwise_selection(X, y, initial_list=[], threshold_in=0.01, threshold_out=0.05, verbose=True):
    """
    Realiza la selección stepwise para encontrar el subconjunto óptimo de predictores.

    Parámetros:
      X - DataFrame con las variables predictoras.
      y - Serie con la variable objetivo.
      initial_list - Lista de variables a incluir inicialmente.
      threshold_in - Umbral para agregar una variable.
      threshold_out - Umbral para eliminar una variable.
      verbose - Si True, imprime cada paso.

    Retorna:
      Lista de variables seleccionadas.
    """
    included = list(initial_list)
    while True:
        changed = False

        # Paso forward: intentar agregar variables
        excluded = list(set(X.columns) - set(included))
        new_pval = pd.Series(index=excluded, dtype=float)
        for new_column in excluded:
            model = sm.OLS(y, sm.add_constant(X[included + [new_column]])).fit()
            new_pval[new_column] = model.pvalues[new_column]
        best_pval = new_pval.min() if not new_pval.empty else None
        if best_pval is not None and best_pval < threshold_in:
            best_feature = new_pval.idxmin()
            included.append(best_feature)
            changed = True
            if verbose:
                print('Agregamos {:30} con p-valor {:.6}'.format(best_feature, best_pval))

        # Paso backward: intentar eliminar variables
        model = sm.OLS(y, sm.add_constant(X[included])).fit()
        # Se excluye el intercepto
        pvalues = model.pvalues.iloc[1:]
        worst_pval = pvalues.max() if not pvalues.empty else None
        if worst_pval is not None and worst_pval > threshold_out:
            worst_feature = pvalues.idxmax()
            included.remove(worst_feature)
            changed = True
            if verbose:
                print('Eliminamos {:30} con p-valor {:.6}'.format(worst_feature, worst_pval))

        if not changed:
            break

    return included
```

```python
import numpy as np
import pandas as pd

# Imprime los tipos actuales para diagnóstico
print("Tipos de datos en X antes de forzar conversión:")
print(X.dtypes)
print("Tipo de y antes de forzar conversión:", y.dtype)

# Forzar la conversión
X = X.apply(lambda col: pd.to_numeric(col, errors='coerce')).astype(float)
y = pd.to_numeric(y, errors='coerce').astype(float)

# Reiniciar el índice para evitar que el índice cause problemas
X = X.reset_index(drop=True)
y = y.reset_index(drop=True)

# Verifica nuevamente
print("Tipos de datos en X después de la conversión:")
print(X.dtypes)
print("Tipo de y después de la conversión:", y.dtype)

# Opcional: verifica el dtype del array subyacente
print("np.asarray(X).dtype:", np.asarray(X).dtype)
print("np.asarray(y).dtype:", np.asarray(y).dtype)
```

```python
X = X.drop(columns=['precio por unidad'], errors='ignore')
```

```python
# Aplicar la selección de variables stepwise nuevamente
selected_features = stepwise_selection(X, y, verbose=True)

print("Variables seleccionadas después de eliminar la colinealidad:", selected_features)

```

```python
# Supongamos que 'df_final' es el DataFrame con todas las transformaciones, eliminación de outliers y selección de variables
X_final = df_final[selected_features]  # Ahora solo usa las variables seleccionadas por Stepwise
y_final = df_final['ventas totales']

# Dividir el dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.2, random_state=42)

# Modelo de Regresión Lineal (para comparar)
from sklearn.linear_model import LinearRegression
modelo_lineal = LinearRegression()
modelo_lineal.fit(X_train, y_train)
y_pred_lineal = modelo_lineal.predict(X_test)

# Evaluar ambos modelos
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

mse_lineal = mean_squared_error(y_test, y_pred_lineal)
rmse_lineal = np.sqrt(mse_lineal)
r2_lineal = r2_score(y_test, y_pred_lineal)

print("Modelo Lineal:")
print("MSE:", mse_lineal,
      " RMSE:", rmse_lineal,
      " R²:", r2_lineal)


```
