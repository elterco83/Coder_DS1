# Modelo de Predicción de Ventas

---

El objetivo de este modelo es predecir las ventas totales de productos Adidas.

Antes de entrenarlo, se elimina la variable *'precio por unidad'*, para evitar que el modelo tenga información redundante y prevenir problemas de colinealidad.
Si la incluyera, el modelo podría depender demasiado de esta relación matemática en lugar de identificar patrones reales en los datos.

Después de ese ajuste, uso el método stepwise para seleccionar automáticamente las variables más importantes para la predicción, descartando las que no aportan mucho.

La técnica que aplicada es Regresión Lineal Múltiple, permitiendo entender cómo influyen distintos factores en las ventas.

La idea es que el modelo no solo haga buenas predicciones, sino que también ayude a tomar mejores decisiones basadas en datos.



**Proceso y Modelado de Datos para Prediccion de Ventas : Paso a Paso**

**1.**   En primer lugar convertimos las variables categoricas a numericas (dummies) , para poder utilizarlas en el modelo de Regresión Lineal Múltiple.

```python
#1

#Se convierten las variables categóricas a dummies
categoricas = ['ciudad','estado','sexo', 'retailer', 'region', 'metodo de venta', 'producto']
df_base = pd.get_dummies(df_base, columns=categoricas, dtype=int)
```

**2.**   En este codigo, se modifica el tipo de dato de las columna 'fecha', para poder crear la columna 'temporada' desde los meses. Luego esa columna tambien la pasamos a dummies (numerica). Por ultimo se elimina la columna fecha, ya que no es necesaria para el modelo. Y se suigue el proceso con df_clean.

```python
#2

# se convierte la columna 'fecha' a datetime
df_base['fecha'] = pd.to_datetime(df_base['fecha'], errors='coerce')

# se determina la temporada a partir del mes
def obtener_temporada(fecha):
    mes = fecha.month
    if mes in [12, 1, 2]:
        return 'Invierno'
    elif mes in [3, 4, 5]:
        return 'Primavera'
    elif mes in [6, 7, 8]:
        return 'Verano'
    elif mes in [9, 10, 11]:
        return 'Otoño'

# Creamos una nueva columna 'temporada' extrayendo la estación de la fecha
df_base['temporada'] = df_base['fecha'].apply(lambda x: obtener_temporada(x) if pd.notnull(x) else None)

# Convertir la variable 'temporada' a dummies y le eliminamos 1 categoria para evitar la multicolinealidad
df_base = pd.get_dummies(df_base, columns=['temporada'], drop_first=True)

# Eliminamos la columna fecha ya que no vamos a necesitarla para el modelo
df_clean = df_base.drop(columns=['fecha'])
```

**3.**   Calculo de metricas estadisticas previo a eliminar los Outliers del dataset, eliminacion de los mismos. Y luego ya sin valores extremos,  se vuelven a calcular la media, mediana y moda.
Se crea 'df_final', como el Dataframe preparado para el modelo.

```python
#3

# Calcular métricas estadísticas sobre df_base
media = df_base.mean()        # Media
mediana = df_base.median()    # Mediana
moda = df_base.mode().iloc[0] # Moda (toma la primera si hay múltiples)

# Mostrar resultados
print("Media con outliers:\n", media, "\n")
print("Mediana con outliers:\n", mediana, "\n")
print("Moda con outliers :\n", moda, "\n")
print("---------" * 8)



# Calculamos los cuartiles y el IQR
Q1 = df_clean['ventas totales'].quantile(0.25)
Q3 = df_clean['ventas totales'].quantile(0.75)
IQR = Q3 - Q1

# Definimos los límites inferior y superior
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Eliminamos lo outliers de df_clean y creamos df_final para usar en el modelo
df_final = df_clean[(df_clean['ventas totales'] >= lower_bound) & (df_clean['ventas totales'] <= upper_bound)]

print("Tamaño original:", df_base.shape)
print("Tamaño después de eliminar outliers:", df_final.shape)
print("---------" * 8)

# Calcular métricas estadísticas sobre df_final
media = df_final.mean()        # Media
mediana = df_final.median()    # Mediana
moda = df_final.mode().iloc[0] # Moda (toma la primera si hay múltiples)

# Mostrar resultados
print("Media sin outliers :\n", media, "\n")
print("Mediana sin outliers :\n", mediana, "\n")
print("Moda sin outliers :\n", moda, "\n")

```

**4.** En este paso, definimos como variables predictoras (X) a todas, excepto a la variable objetivo (y).

```python
#4

X = df_final.drop(columns=['ventas totales']) # variables predictorias
y = df_final['ventas totales'] # variable objetivo
```

**5.**  Luego se fuerza la conversion de todas las variables a float64 para evitar errores en el modelo y reinicio los índices para mantener alineados X e y, asegurando un dataset limpio y listo para el entrenamiento

```python
#5

# Imprime los types actuales para diagnóstico
print("Tipos de datos en X antes de forzar conversión:")
print(X.dtypes)
print("Tipo de y antes de forzar conversión:", y.dtype)

# Forzar la conversión
X = X.apply(lambda col: pd.to_numeric(col, errors='coerce')).astype(float)
y = pd.to_numeric(y, errors='coerce').astype(float)

# Reiniciar el índice para evitar que el índice cause problemas
X = X.reset_index(drop=True)
y = y.reset_index(drop=True)

# Verifica nuevamente
print("Tipos de datos en X después de la conversión:")
print(X.dtypes)
print("Tipo de y después de la conversión:", y.dtype)
```

**6.** Se aplica la selección de variables stepwise , las misma prepara un sub-conjunto de predictores en función de su p-valor.
Asegurando un modelo más eficiente con solo las variables más importantes.

```python
#6

def stepwise_selection(X, y, initial_list=[], threshold_in=0.01, threshold_out=0.05, verbose=True):
    """
    Selección stepwise para encontrar el subconjunto óptimo de predictores.

    Parámetros:
      X - DataFrame con las variables predictoras.
      y - Serie con la variable objetivo.
      initial_list - Lista de variables a incluir inicialmente.
      threshold_in - Umbral para agregar una variable.
      threshold_out - Umbral para eliminar una variable.
      verbose - Si True, imprime cada paso.

    Retorna:
      Lista de variables seleccionadas.
    """
    included = list(initial_list)
    while True:
        changed = False

        # Paso forward: intentar agregar variables
        excluded = list(set(X.columns) - set(included))
        new_pval = pd.Series(index=excluded, dtype=float)
        for new_column in excluded:
            model = sm.OLS(y, sm.add_constant(X[included + [new_column]])).fit()
            new_pval[new_column] = model.pvalues[new_column]
        best_pval = new_pval.min() if not new_pval.empty else None
        if best_pval is not None and best_pval < threshold_in:
            best_feature = new_pval.idxmin()
            included.append(best_feature)
            changed = True
            if verbose:
                print('Agregamos {:30} con p-valor {:.6}'.format(best_feature, best_pval))

        # Paso backward: intentar eliminar variables
        model = sm.OLS(y, sm.add_constant(X[included])).fit()
        # Se excluye el intercepto
        pvalues = model.pvalues.iloc[1:]
        worst_pval = pvalues.max() if not pvalues.empty else None
        if worst_pval is not None and worst_pval > threshold_out:
            worst_feature = pvalues.idxmax()
            included.remove(worst_feature)
            changed = True
            if verbose:
                print('Eliminamos {:30} con p-valor {:.6}'.format(worst_feature, worst_pval))

        if not changed:
            break

    return included

```

**7.**   Antes de ejecutar el modelo stepwise, se elimina la variable *'precio por unidad'*, para evitar que el modelo tenga información redundante y prevenir problemas de colinealidad.

```python
#7

#se elimina una variable predictoria de X antes de las seleccion
X = X.drop(columns=['precio por unidad'], errors='ignore')
```

**8.**  Se ejecuta stepwise_selection sobre X e y, dando como resultado las variables seleccionadas para ejecutar el modelo de regresion lineal multiple.

```python
#8

# Ejecutamos la selección de variables stepwise
selected_features = stepwise_selection(X, y, verbose=True)

print("Variables seleccionadas:", selected_features)

```

**9.**   Para la ejecucion de codigo, almacenamos X (variables seleccionadas) e y (variable objetivo) código, luego se dividen los datos para entrenar y probar al modelo y ya con esa informacion pasamos a  entrenar un modelo de regresión lineal.

```python
#9

#importacion de librerias para el entrenamiento del modelo

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# preparamos X e y para el modelo
X_final = df_final[selected_features]
y_final = df_final['ventas totales']

# Dividimos X para entrenar y probar implementando la semilla
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.2, random_state=42)

# Modelo de Regresión Lineal (para comparar)
modelo_lineal = LinearRegression()
modelo_lineal.fit(X_train, y_train)
y_pred_lineal = modelo_lineal.predict(X_test)


```

**10.** El ultimo paso, evaluamos el modelo con métricas como MSE, RMSE y R².
Tambien calculasmos que porcetaje representa el RSME en el promedio de ventas totales.

```python
#10

# Importacion de libreria para la Evaluacion del modelo
from sklearn.metrics import mean_squared_error, r2_score

mse_lineal = mean_squared_error(y_test, y_pred_lineal)
rmse_lineal = np.sqrt(mse_lineal)
r2_lineal = r2_score(y_test, y_pred_lineal)

print("Modelo de Regresion Lineal:")
print(f"MSE en dolares al cuadrado: {mse_lineal} USD²")
print(f" RMSE en dolares: {rmse_lineal} USD")
print("R²:" ,r2_lineal)

#calculo del promedio de ventas
promedio_ventas = y.mean()
print("Promedio de ventas (en USD ):", promedio_ventas / 1000)

#porcentaje de rsme sobre el pomedio de ventas
rmse_lineal = np.sqrt(mse_lineal)
error_relativo = (rmse_lineal / promedio_ventas) * 100
print("Porcentaje de error relativo:", error_relativo, "%")
```
